```python
# import module
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_current_air_pollution_data(appid, latitude, longitude):
    url = f'http://api.openweathermap.org/data/2.5/air_pollution?lat={latitude}&lon={longitude}&appid={appid}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print('Failed to retrieve current air pollution data.')
        return None


api_key = '9eac0c802aa6ff1b0ee6a5d811af44d0'
latitude = 50
longitude = 50

current_data = get_current_air_pollution_data(api_key, latitude, longitude)
```

```python
def get_forecast_air_pollution_data(appid, latitude, longitude):
    url = f'http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={latitude}&lon={longitude}&appid={appid}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print('Failed to retrieve forecast air pollution data.')
        return None


api_key = '9eac0c802aa6ff1b0ee6a5d811af44d0'
latitude = 50
longitude = 50

forecast_data = get_forecast_air_pollution_data(api_key, latitude, longitude)
```

```python
forecast_data
```

```python
def get_historical_air_pollution_data(appid, latitude, longitude,start,end):
    url = f'http://api.openweathermap.org/data/2.5/air_pollution/history?lat={latitude}&lon={longitude}&start={start}&end={end}&appid={appid}'
#     f'http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={latitude}&lon={longitude}&appid={appid}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print('Failed to retrieve forecast air pollution data.')
        return None


api_key = '9eac0c802aa6ff1b0ee6a5d811af44d0'
latitude = 50
longitude = 50
start = 1356998400
end = 1672531200

historic_data = get_historical_air_pollution_data(api_key, latitude, longitude,start,end)
```

```python
merged_data_new = {
    'coord': current_data['coord'],
    'list': current_data['list'] + forecast_data['list'] + historic_data['list']
}
```

```python
df2 = pd.json_normalize(merged_data_new['list'])
```

```python
df_s = df2.sort_values('dt', ascending=True).reset_index(drop = True)
```

```python
df_s
```

```python
x = df_s.drop(['dt','main.aqi'],axis = 1)
y = df_s['main.aqi']
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize the XGBoost regressor
model = xgb.XGBRegressor()

# Train the model
model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)

print("Root Mean Squared Error:", rmse)
```

```python
from sklearn.metrics import r2_score

r2_s = r2_score(Y_test,Y_pred)
r2_s
```
